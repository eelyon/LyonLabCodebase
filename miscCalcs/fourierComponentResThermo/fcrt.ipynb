```python
import numpy as np
from scipy. fft import fft, fftfreq
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
fm = int(6.21 * (1e3))
fc = int(100 * (1e3))
A = 1 * 1/2

t0 = 3e-4
npc = 100
numPoints = int(npc * t0 * fc)

tvec = np.linspace(0, t0, num = numPoints)
sig = (A / 1) * np.sin(2 * np.pi * fc * tvec) * (1 + np.sin(2 * np.pi * fm * tvec))
```

```python
plt.plot(tvec * 1000, sig)
plt.xlabel("Time [ms]")
plt.ylabel("Amplitude [a.u.]")

SMALL_SIZE = 20
MEDIUM_SIZE = 16
BIGGER_SIZE = 30

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize

#plt.savefig("heatingSignal_2025_03_06.png", bbox_inches = 'tight', pad_inches = 0.1, dpi = 500)
plt.show()
```

```python
sigtrans = fft(sig)
sigsqtrans = fft(np.power(sig, 2))
sf = 2.0/numPoints * np.abs(sigtrans[0:numPoints//2])
ssf = 2.0/numPoints * np.abs(sigsqtrans[0:numPoints//2])
xf = fftfreq(numPoints, t0 / numPoints)[0:numPoints//2]
```

```python
plt.plot(xf, np.real(sf))
plt.xlim(0,1.1 * fc)

plt.show()
```

```python
plt.plot(xf, np.real(ssf))
plt.xlim(-0,2.2 * fm)
plt.show()
```
