```python
import numpy as np
from scipy. fft import fft, fftfreq
import matplotlib as mlp
import matplotlib.pyplot as plt
```

```python
fm = int(3.6 * (1e3))
fc = int(100 * (1e3))
A = 0.01

t0 = 0.1
npc = 10
numPoints = int(npc * t0 * fc)

tvec = np.linspace(0, t0, num = numPoints)
sig = (A / 1) * np.sin(2 * np.pi * fc * tvec) * (1 + np.sin(2 * np.pi * fm * tvec))
```

```python
plt.plot(tvec, sig)
plt.show()
```

```python
sigtrans = fft(sig)
sf = 2.0/numPoints * np.abs(sigtrans[0:numPoints//2])
xf = fftfreq(numPoints, t0 / numPoints)[0:numPoints//2]
```

```python
plt.plot(xf, np.real(sf))
plt.xlim(0,1.1 * fc)

plt.show()
```

```python
plt.plot(xf, np.imag(sf))
plt.xlim(0,1.1 * fc)
plt.show()
```
